```python
def table_transformer(df):
    df_copy = df.copy()
    df_type = df_copy.loc[df_copy['Type Value'].notnull()]
    #df_type.reset_index().drop(['index'], axis=1, inplace=True)
    df_type['AttrIsValueOrType'] = 'type'
    df_type.reset_index(drop=True)
    df_type.drop(['Attributes Value', 'Результат', 'fileName', 'Unnamed: 7', 'Дата', 'Частота' ], axis=1, inplace=True)
    df_type.rename(columns={'Type Value': 'Attributes Value'}, inplace=True)
    
    # т.к. копию всех записей где задан тип атрибута уже сделали и подготовили для мёрджа, обработаем исходный датафрейм
    df_copy['AttrIsValueOrType'] = 'value' # зададим для всех строк тип value (для тех где не уакзан Type Value и для тех где указан, но мы уже перенесли в отдельный и уже отметили как type )
    df_copy.drop(['Type Value', 'Результат', 'fileName', 'Unnamed: 7', 'Дата', 'Частота' ], axis=1, inplace=True) # грохнет type Value, т.к. его уже переложили в отдельный датафрейм и все остальные
    
    full_df = pd.concat([df_copy, df_type], axis = 0)
    full_df.reset_index(drop=True)
    
    full_df.sort_values(by=['AccountId', 'GTIN'])
    
    return full_df
    
```

```python
#AttrIsValueOrType = 'value'
#AccountId = '155183'
#gtin = '4670016810069'
#AttrId = '2716'

#df = pd.DataFrame()

def DB_extractor(AttrIsValueOrType, AccountId, gtin, AttrId ):
    dfl = pd.DataFrame()
    
    con = pymysql.connect(host=host, user=user, port=port, password=password, database=database, cursorclass=pymysql.cursors.DictCursor)
    with con:
        cur = con.cursor()

        query = "SET @AttrIsValueOrType = \'{}\'".format(AttrIsValueOrType)
        cur.execute(query)

        query = '''SELECT lg.MainAccountId, lg.Barcode, lg.GoodId,  latga.MarkGroupId, lgtc.CatId, lav.AttrId
            ,latga.GS1AttrId
            ,lav.Value, lav.`Type` 
            ,(CASE WHEN @AttrIsValueOrType = 'value' THEN lav.Value WHEN @AttrIsValueOrType = 'type' THEN lav.Type END) as comparedValue
            ,latga.TypeCondition
            ,latga.AttrIsValueOrType -- вспомогательный
            ,latga.DefaultType
            ,latga.GS1AttrName 
            ,latga.GS1AttrIsDictionary -- НЕ обязательный
            FROM Lst_AttrValues lav 
            JOIN Lst_Goods lg ON lav.GoodId = lg.GoodId  AND (lg.Status = 'published' or lg.Status = 'archived') AND lg.Deleted = 0
            LEFT JOIN Lst_GoodsToCat lgtc ON lgtc.GoodId = lg.GoodId  AND lgtc.Deleted = 0 
            LEFT JOIN Lst_AttrToGS1Attr latga ON latga.CatId =  lgtc.CatId AND latga.AttrId = lav.AttrId  AND latga.Operation = 'export' AND latga.GS1AttrId NOT LIKE '%BRICK'   AND latga.Deleted = 0  
                 AND (CASE WHEN (lav.Type = 'Артикул' OR  lav.Type = 'Модель') THEN latga.TypeCondition = lav.Type ELSE (latga.TypeCondition <=> NULL OR latga.TypeCondition = '') END)
            WHERE lav.Deleted = 0   AND AttrIsValueOrType = @AttrIsValueOrType
            AND lg.MainAccountId = {}
            AND lg.Barcode = '{}' 
            AND lav.AttrId = {}; '''.format(AccountId, gtin, AttrId)
        cur.execute(query)

        #query = 'SELECT @AttrIsValueOrType'
        #cur.execute(query)

        DBresponse = cur.fetchall()


    if len(DBresponse) != 0:

        for i, row in  enumerate(DBresponse):
            dfl.loc[i,'MainAccountId'], dfl.loc[i,'Barcode'], dfl.loc[i,'GoodId'], dfl.loc[i,'MarkGroupId'], dfl.loc[i,'CatId'], \
            dfl.loc[i,'AttrId'], dfl.loc[i,'GS1AttrId'], dfl.loc[i,'Value'], dfl.loc[i,'Type'], dfl.loc[i,'comparedValue'], \
            dfl.loc[i,'TypeCondition'], dfl.loc[i,'AttrIsValueOrType'], dfl.loc[i,'DefaultType'], dfl.loc[i,'GS1AttrName'], dfl.loc[i,'GS1AttrIsDictionary'] = \
            row['MainAccountId'], row['Barcode'], row['GoodId'], row['MarkGroupId'], row['CatId'], \
            row['AttrId'], row['GS1AttrId'], row['Value'], row['Type'], row['comparedValue'], \
            row['TypeCondition'], row['AttrIsValueOrType'], row['DefaultType'], row['GS1AttrName'], row['GS1AttrIsDictionary'] 

    else:
        print('вошли в else')
        dfl.loc[0,'MainAccountId'], dfl.loc[0,'Barcode'],  dfl.loc[0,'AttrId'],  dfl.loc[0,'AttrIsValueOrType'] = AccountId, gtin, AttrId, AttrIsValueOrType

    return dfl    
```
